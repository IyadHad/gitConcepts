```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,3*np.pi,500)#500 = nombres de valeures prises
y=x**2
print(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('f(x)')
#plt.xlim(-1.5,1.5)
y2 = 32*x
plt.plot(x,y2)
plt.title("graph of x^2 vs X")
plt.legend()
plt.show()
```
